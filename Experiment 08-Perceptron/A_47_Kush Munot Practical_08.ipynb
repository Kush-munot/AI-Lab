AND GATE

```python
w = [1,1]
x = [[0,0],
     [0,1],
     [1,0],
     [1,1]]
b = []
y = []

def and_gate():
  for i in range(0,4):
    if (1 in x[i]):
      b.append(-1)
    else:
      b.append(0)
    y.append(w[0]*x[i][0] + w[1]*x[i][1] + b[i])
  return y

print("Values Predicted for AND gate:")
output = and_gate()
for i in range(0,4):
  print("{} AND {} = {}".format(x[i][0],x[i][1],output[i]))
```

OR GATE

```python
INPUT=[(0,0),(0,1),(1,0),(1,1)]
w1,w2=1,1
def ORGATE(ins):
	b=None
	output=[]
	for i in ins:
		if(i[0]==0 or i[1]==0):
			b=0
		else:
			b=-1
		output.append(w1*i[0]+w2*i[1]+b)
	return output


print("Values Predicted for OR gate:")
output=ORGATE(INPUT)
t=0
for i in INPUT:
	print("{} OR {} = {}".format(i[0],i[1],output[t]))
	t+=1
```

NAND GATE

```python
inputs=[[0,0],[0,1],[1,0],[1,1]]
W=[1,1]
bias=[]
def nand_gate():
  for i in range(0,4):
    if(inputs[i][0]==1 and inputs[i][1]==1):
      bias.append(-2)
    elif(inputs[i][0]==0 and inputs[i][1]==0):
      bias.append(1)
    else:
      bias.append(0)
  Y=[]
  for j in range(0,4):
    Y.append((W[0]*inputs[j][0])+(W[1]*inputs[j][1])+bias[j])
  return Y
print("Values predicted for NAND Gate:")
N=nand_gate()
for j in range(0,4):
  print("{} NAND {}= {}".format(inputs[j][0],inputs[j][1],N[j]))
```
